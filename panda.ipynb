```python
df_sco=pd.read_csv(r"C:\Users\udaya\Downloads\deliveries.csv")
```

```python
df_run=df_sco.groupby(['match_id','batsman']).apply(lambda x:np.sum(x['batsman_runs'])).reset_index(name='Innings_Runs')
```

```python
plt.figure(figsize=(15,8))
plot4=sns.displot(df_run['Innings_Runs'],kde = False)
plt.title("Histograms for runs scored",fontsize=20)
plt.xticks(rotation="vertical")
plt.ylabel("number of times")
plt.xlabel("runs scored")
plt.show()
```

```python
df_run=df_sco.groupby(['match_id','batsman']).apply(lambda x : np.sum(x['batsman_runs'])).reset_index(name='Innings_Runs')
df_faced_ball = df_sco.groupby(['match_id','batsman']).apply(lambda x : x['batsman_runs'].count()).reset_index(name='Balls_Faced')
df_run_ball = df_faced_ball.merge(df_run[['match_id','batsman','Innings_Runs']])
scatter_ball_runs=df_run_ball[df_run_ball['Balls_Faced'] > 20]

```

```python
plt.figure(figsize=(15,8))
plot5=sns.scatterplot(x='Balls_Faced',y='Innings_Runs',data=scatter_ball_runs)
plt.title("Runs Scored vs Balls Faced",fontsize=20)
plt.xlabel("Ball Faced")
plt.ylabel("Runs Scored")
plt.show()

```
