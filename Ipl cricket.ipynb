```python
#top player of the match winner
top_player=df_match.player_of_match.value_counts()[:20]
fig ,ax=plt.subplots(figsize=(15,8))
ax .set_ylim([0,40])
ax .set_ylabel("count")
ax.set_title("top player of the match winners")
top_player.plot.bar()
sns.barplot(x=top_player.index,y=top_player,orient='v',palette='Reds')
plt.show()
```

```python
#total toss decision percentagematch
temp_series=df_match.toss_decision.value_counts()
labels=(np.array(temp_series.index))
sizes=(np.array((temp_series/temp_series.sum())*100))
colors=['red','blue']
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True,startangle=90)
plt.title("Toss Decision percentage")
plt.show()
```

```python
#season task decision using hue-bar graph
plt.figure(figsize=(12,4))
sns.countplot(x='season', hue='toss_decision',data=df_match)
plt.xticks(rotation='vertical')
plt.show()
```

```python
#top run score in overall season
df_temp=df_deliver.groupby('batsman')['batsman_runs'].agg('sum')
df_temp
```

```python
df_temp=df_deliver.groupby('batsman')['batsman_runs'].agg('sum').reset_index()
df_temp
```

```python
df_temp=df_deliver.groupby('batsman')['batsman_runs'].agg('sum').reset_index().sort_values(by='batsman_runs',ascending=False)
df_temp
```

```python
df_temp=df_deliver.groupby('batsman')['batsman_runs'].agg('sum').reset_index().sort_values(by='batsman_runs',ascending=False).reset_index(drop=True)
df_temp
```

```python
df_temp=df_deliver.groupby('batsman')['batsman_runs'].agg('sum').reset_index().sort_values(by='batsman_runs',ascending=False).reset_index(drop=True)
df_temp
```

```python
#players with boundary in IPL
df_temp=df_deliver.groupby('batsman')['batsman_runs'].agg(lambda x:(x==4).sum()).reset_index().sort_values(by='batsman_runs',ascending=False).reset_index(drop=True)
df_temp
```

```python
#players with 6 in IPL
df_temp=df_deliver.groupby('batsman')['batsman_runs'].agg(lambda x:(x==6).sum()).reset_index().sort_values(by='batsman_runs',ascending=False).reset_index(drop=True)
df_temp
```

```python
#players with 0 in IPL
df_temp=df_deliver.groupby('batsman')['batsman_runs'].agg(lambda x:(x==0).sum()).reset_index().sort_values(by='batsman_runs',ascending=False).reset_index(drop=True)
df_temp
```

```python
#no of balls bowled by eacg player
df_temp=df_deliver.groupby('bowler')['ball'].agg('count').reset_index().sort_values(by='ball',ascending=False).reset_index(drop=True)
df_temp
```

```python
#no of balls not attended by boler
df_temp=df_deliver.groupby('bowler')['total_runs'].agg(lambda x: (x==0).sum()).reset_index().sort_values(by='total_runs',ascending=False).reset_index(drop=True)
df_temp
```

```python
#extra runs
df_temp=df_deliver.groupby('bowler')['extra_runs'].agg(lambda x: (x>0).sum()).reset_index().sort_values(by='extra_runs',ascending=False).reset_index(drop=True)
df_temp
```
